```python
import numpy as np #Linear Algebra
import pandas as pd #Data processing, CSV File I/O (e.g read_csv)
import seaborn as sbn #Data visualization using seaborn
import matplotlib.pyplot as plt #Plot 
import time
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
col = data.columns
print(col)
```

# CLEANING UNUSED DATA

```python
y = data.diagnosis #Save column diagnosis to analysis or target variable
drop_cols = ['Unnamed: 32','id','diagnosis'] #Define columns to drop or delete
x = data.drop(drop_cols, axis=1)
x.head()
```

# PLOT DIAGNOSIS DISTRIBUTIONS

```python
B, M = y.value_counts()
plt.figure(figsize=(10,9))
sbn.set_theme()
line1, =plt.bar('Benign Tumors', B)
line2, =plt.bar('Malignant Tumors', M)
line1.set_label('Benign Tumors')
line2.set_label('Malignant Tumors')
plt.legend()
plt.show()
```

```python
ax = sbn.countplot(y, label="Count") #Plot diagnosis tumors
B, M = y.value_counts() #Count individual diagnosis variable of tumors

print('Number of Benign Tumors', B) 
print('Number of Malignant Tumors', M)
```

```python
x.describe() #To see features of data
```

# DATA VISUALIZATION

### VISUALIZING STANDARDIZED DATA WITH SEABORN

```python
data = x #Features
data_std = (data - data.mean())/data.std() #To define standardized data
data = pd.concat([y, data_std.iloc[:]], axis=1) #Merge string and integer data with pd.concat
data = pd.melt(data, id_vars='diagnosis',
              var_name='Features',
              value_name='Value')

plt.figure(figsize=(20,20))
sbn.violinplot(x='Features', y='Value', hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45)
```

### VIOLIN PLOTS AND BOX PLOTS

```python
data = pd.concat([y, data_std.iloc[:, 20:34]], axis=1) #Merge string and integer data with pd.concat #Analyze 20th to 34th data
data = pd.melt(data, id_vars='diagnosis',
              var_name='Features',
              value_name='Value')

plt.figure(figsize=(20,20))
sbn.violinplot(x='Features', y='Value', hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45)
```

```python
sbn.boxplot(x='Features', y='Value', hue='diagnosis', data=data)
plt.xticks(rotation=45);
```

### USING JOINT PLOTS TO FEATURE COMPARISON

```python
sbn.jointplot(x.loc[:, 'concavity_worst'],
             x.loc[:, 'concave points_worst'],
             kind='reg',
             color='#ce1414');
```

### OBSERVING THE DISTRIBUTION OF VALUES AND THEIR VARIANCE WITH SWARM PLOTS

```python
sbn.set(style='whitegrid', palette='muted')
data = x #Features
data_std = (data - data.mean())/data.std() #To define standardized data
data = pd.concat([y, data_std.iloc[:, 0:14]], axis=1) #Merge string and integer data with pd.concat
data = pd.melt(data, id_vars='diagnosis',
              var_name='Features',
              value_name='Value')

plt.figure(figsize=(10,10))
sbn.swarmplot(x='Features', y='Value', hue='diagnosis', data=data)
plt.xticks(rotation=45)
```
